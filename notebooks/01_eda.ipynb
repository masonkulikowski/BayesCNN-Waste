```python
import sys
sys.path.append('..')

from src.config import load_config
from src.load_data import load_data, get_class_dist
import matplotlib.pyplot as plt
```

```python
config = load_config()
print(f"Project: {config['project']['name']}")
print(f"Dataset: {config['data']['dataset_name']}")
print(f"Image size: {config['data']['image_size']}x{config['data']['image_size']}")

full_ds = load_data(split_data=False)
print(f"\nTotal Images: {len(full_ds)}")

freq_dict = get_class_dist(full_ds)

plt.figure(figsize=(10, 6))
plt.bar(freq_dict.keys(), freq_dict.values(), color='steelblue', edgecolor='black')
plt.xlabel('Waste Category', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)
plt.title('Class Distribution in TrashNet Dataset', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nClass Distribution:")
print("-" * 40)
for label, count in freq_dict.items():
    percentage = (count / sum(freq_dict.values())) * 100
    print(f"{label:12s}: {count:4d} images ({percentage:5.2f}%)")
print("-" * 40)
print(f"{'Total':12s}: {sum(freq_dict.values()):4d} images")
```
